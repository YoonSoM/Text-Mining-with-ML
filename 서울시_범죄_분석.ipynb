#2. 범죄 발생 횟수

우리는 서울시의 구별 범죄 데이터를 통해 범죄의 발생 자체가 적거나 혹은

범죄가 발생했더라도 검거가 잘되는 일명 '안전한' 구가 어딘지 알아 볼 수 있습니다. (그 반대도 가능합니다) 

## 1. 가장 많이 일어나는 지역과 가장 적게 일어나는 지역

그럼 단순히 전체 범죄의 수로만 따졌을때 

가장 범죄가 많이 일어나는 서울시 구와 가장 범죄가 적게 일어나는 서울시 구를 각각 5개씩 찾아보겠습니다.

##### 문제13. 범죄가 가장 많이 일어났던 구와 적게 일어났던 구를 5개씩 출력하세요. 

```python
years = ['2014','2015','2016','2017','2018','2019']

high ={}
low = {}

for year in years:
    data = crime_data[crime_data['기간'] == year]
    data= data.sort_values(by='총 범죄', ascending=False)

    high_tmp = [gu for gu in data['자치구'][:5]]
    
    for ht in high_tmp:
        if ht not in high:
            high[ht] =1
        else:
            value = high[ht]
            high[ht] =value +1

    low_tmp = [gu for gu in data['자치구'][-5:]]
    
    for lt in low_tmp:
        if lt not in low:
            low[lt] =1
        else:
            value = low[lt]
            low[lt] =value +1
```

```python
import operator

high= sorted(high.items(), key=operator.itemgetter(1), reverse =True)
low= sorted(low.items(), key=operator.itemgetter(1), reverse =True)

print(high)
print()
print(low)
```

```python
high5 = [h for h,_ in high[:5]]
low5 = [l for l,_ in low[:5]]

print('가장 범죄가 많이 일어나는 서울시 TOP 5구')
print(high5, end='\n\n')

print('가장 범죄가 적게 일어나는 서울시 TOP 5구')
print(low5)
```

```python
years = ['2014','2015','2016','2017','2018','2019']

weak ={}
strong = {}

for year in years:
    data = crime_data[crime_data['기간'] == year]
    data= data.sort_values(by='총 검거율', ascending=False)

    weak_tmp = [gu for gu in data['자치구'][-5:]]
    
    for wt in weak_tmp:
        if wt not in weak:
            weak[wt] =1
        else:
            value = weak[wt]
            weak[wt] =value +1

    strong_tmp = [gu for gu in data['자치구'][:5]]
    
    for st in strong_tmp:
        if st not in strong:
            strong[st] =1
        else:
            value = strong[st]
            strong[st] =value +1
```

```python
weak= sorted(weak.items(), key=operator.itemgetter(1), reverse =True)
strong= sorted(strong.items(), key=operator.itemgetter(1), reverse =True)

print(weak, end='\n\n')
print(strong)
```

```python
strong5 = [s for s,_ in strong[:5]]
weak5 = [w for w,_ in weak[:5]]

print('가장 검거율이 높은 서울시 TOP 5구')
print(strong5, end='\n\n')

print('가장 검거율이 낮은 서울시 TOP 5구')
print(weak5)
```
